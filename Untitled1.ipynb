```python
class Vector:
    def __init__(self, l):
        self.l=l
        
    def as_list(self):
        return list(self.l)
    
    def size(self):
        return len(self.l)
    
    def magnitude(self):
        return abs(sum([a**2 for a in self.l])**(1/2))

    def euclidean_distance(self, other):
        if isinstance(other, Vector):
            return (sum([(a-b)**2 for a,b in zip(self.l, other.as_list())]))**0.5
        return None
    
    def normalized(self):
        return Vector([a/self.magnitude() for a in self.l])
    
    def __add__(self, other):
        if isinstance(other, Vector) and other.size() == self.size():
            return Vector([a+b for a,b in zip(self.l, other.as_list())])
        return None
    
    def __sub__(self, other):
        if isinstance(other, Vector) and other.size() == self.size():
            return Vector([a-b for a,b in zip(self.l, other.as_list())])
        return None
    
    def __mul__(self, other):
        if isinstance(other, Vector) and other.size() == self.size():
            return sum([a*b for a,b in zip(self.l, other.as_list())])
        if isinstance(other, (int, float)):
            return Vector([a*other for a in self.l])
        return None
    
    def __truediv__(self, other):
        if isinstance(other, (int, float)):
            return Vector([a/other for a in self.l])
        return None
```

```python
def sample_mean(values):
    return sum(values)/len(values)

def sample_var(values):
    mean = sample_mean(values)
    return 1/(len(values)-1) * sum([(a - mean)**2 for a in values])

def sample_std(values):
    return sample_var(values)**0.5

def sample_cov(x_values, y_values):
    if len(x_values) != len(y_values):
        return None
    x_mean = sample_mean(x_values)
    y_mean = sample_mean(y_values)
    return 1/(len(x_values)-1) * sum([(a - x_mean)*(b - y_mean) for a,b in zip(x_values, y_values)])


def pearson_r(x_values, y_values):
    return sample_cov(x_values, y_values)/(sample_std(x_values)*sample_std(y_values))

def linear_regression(x_values, y_values):
    m = sample_cov(x_values, y_values)/sample_var(x_values)
    b = sample_mean(y_values) - m*sample_mean(x_values)
    
    return m, b, pearson_r(x_values, y_values)
```

```python
from scipy import optimize
import math

def collect_vars(constraints):
    names=list(set([element[1] for constraint in constraints for element in constraint[0]]))
    return {names[i]:i for i in range(len(names))}

def solve_linprog(obj, constraints, bounds, minimize):

    name_dict = collect_vars(constraints)

    bounds = [bounds[var] for var in name_dict]
    c=[0 for i in range(len(name_dict))]
    for val in obj:
        c[name_dict[val[1]]]= val[0]

    if minimize == False:
        c = [-1 * element for element in c]
    
    A = [[0 for i in range(len(name_dict))] for j in range(len(constraints))]
    B = [0 for i in range(len(constraints))]
    A_eq=[]
    B_eq=[]
    
    if len(constraints)<1:
        A=None
        b=None
    
    for constraint in range(len(constraints)):
        for var in name_dict:
            var_index = [val for val in range(len(constraints[constraint][0])) 
                         if constraints[constraint][0][val][1] == var]
            if var_index == []:
                continue
            var_index = var_index[0]

            A[constraint][name_dict[var]] = constraints[constraint][0][var_index][0]
        B[constraint]=constraints[constraint][2]
        if constraints[constraint][1] == ">=":
            B[constraint] = -1 * B[constraint]
            A[constraint] = [i * -1 for i in A[constraint]]
        elif constraints[constraint][1] == "=":
           #A.append([i * -1 for i in A[constraint]])
            #B.append(-1 * B[constraint])
            A_eq.append(A[constraint])
            B_eq.append(B[constraint])
    
    if B_eq:
        solution = optimize.linprog(c=c, A_ub=A, b_ub=B, A_eq=A_eq, b_eq=B_eq, bounds=bounds, options={'maxiter':100000})
    else:
        solution = optimize.linprog(c=c, A_ub=A, b_ub=B, bounds=bounds, options={'maxiter':100000})
    opt_values = {}
    
    if isinstance(solution['x'], list): # Check for Nan
        return None
    
    for key in name_dict:
        multiply = -1 if minimize==False else 1
        opt_values[key] = solution['x'][name_dict[key]]
        
    return opt_values, solution.fun * multiply
```

```python
import csv

def solve_transshipment(supply, customers, edges):
    obj =[(edges[a], a) for a in edges]
    constraints = []
    
    for s in supply:
        constraints.append([[(1, e) for e in edges if s in e], "<=", supply[s]])
        
    for c in customers:
        constraints.append([[(1, e) for e in edges if c in e], "=", customers[c]])
 
    internal_nodes = []
    for e in edges:
        if e[0] not in supply and e[0] not in customers:
            internal_nodes.append(e[0])
        if e[1] not in supply and e[1] not in customers:
            internal_nodes.append(e[1])
    internal_nodes = list(set(internal_nodes))

    # Zero sum for all internal nodes
    for node in internal_nodes:
        constraints.append([[(-1, e) if node == e[0]
                             else (1, e)
                             if node == e[1] else (0, e)
                             for e in edges if node in e], "=", 0])
        
    bounds = {}
    for edge in edges:
        bounds[edge] = (0, None)
    #print(obj)
    #print(constraints)
    #print(bounds)
    print("constraints")
    print(len(constraints))
    return solve_linprog(obj, constraints, bounds, True)
```

```python
from scipy import optimize
import numpy as np
import math

def collect_vars(constraints):
    names=list(set([element[1] for constraint in constraints for element in constraint[0]]))
    return {names[i]:i for i in range(len(names))}

def solve_dual(obj, const, minimize, y_names):
    
    constraints = const

    name_dict = collect_vars(constraints)

    # The code below is the same as the LP code (changes at the end)
    c=[0 for i in range(len(name_dict))]
    for val in obj:
        c[name_dict[val[1]]]= val[0]

    if minimize == False:
        c = [-1 * element for element in c]
    
    A = [[0 for i in range(len(name_dict))] for j in range(len(constraints))]
    B = [0 for i in range(len(constraints))]
    A_eq=[]
    B_eq=[]
    
    if len(constraints)<1:
        A=None
        b=None
    
    for constraint in range(len(constraints)):
        for var in name_dict:
            var_index = [val for val in range(len(constraints[constraint][0])) 
                         if constraints[constraint][0][val][1] == var]
            if var_index == []:
                continue
            var_index = var_index[0]

            A[constraint][name_dict[var]] = constraints[constraint][0][var_index][0]
        B[constraint]=constraints[constraint][2]
        if constraints[constraint][1] == ">=" :
            B[constraint] = -1 * B[constraint]
            A[constraint] = [i * -1 for i in A[constraint]]
        elif constraints[constraint][1] == "=":
            A.append(A[constraint])
            B.append(B[constraint])

    # We just need to take the transpose for the dual
    A = [[-y for y in x] for x in list(map(list, zip(*A)))]
    b = [x for x in B]
    c = [x for x in c]

    bounds = [(0, None) for var in b]
    solution = optimize.linprog(c=b, A_ub=A, b_ub=c, bounds=bounds, options={'maxiter':100000})
    opt_values = {}
    
    if isinstance(solution['x'], list): # Check for Nan
        return None
    
    for i in range(len(y_names)):
        #multiply = -1 if minimize==False else 1
        opt_values[y_names[i]] = solution['x'][i]
        
    return opt_values, solution.fun #* multiply
```
