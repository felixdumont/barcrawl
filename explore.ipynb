```python
import numpy as np
import pandas as pd
import json
```

### Import dataset

```python
df = pd.read_json('data/yelp_dataset/business.json', lines=True)
```

```python
df['categories'] = df['categories'].astype('str') 
```

```python
# Only keep bars
def getbars(bars):
    df_bars = bars[(df['categories'].str.contains('Bars') 
                 | df['categories'].str.contains('Nightlife') 
                 | df['categories'].str.contains('Pubs') )
                 & ( ~df['categories'].str.contains('Sushi Bars') 
                 & ~df['categories'].str.contains('Juice Bars')) ]
    return df_bars
```

```python
#only keep specified city
def citybars(bars, city):
    bars_city = bars[(bars['city'].str.contains(city))]
    return bars_city

#only keep open bars
def openbars(bars):
    open_bars = bars[(bars['is_open']==1)]
    return open_bars
```

```python
def hoursbyday(bars):
    #separate hours into days
    bars=citybars(openbars(bars), "Toronto")
    hours=bars['hours'].apply(pd.Series)
    bars=pd.concat([bars, hours],axis=1).drop('hours', axis=1)

    #separate days into open/close times
    days=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    for day in days:
        temp=bars[day].str.split("-", n=1, expand=True)
        bars[day+' open']=temp[0]
        bars[day+' close']=temp[1]
    return bars

def separate_attributes(bars):
    attributes=bars['attributes'].apply(pd.Series)
    bars=pd.concat([bars, attributes],axis=1).drop('attributes', axis=1)


#keep only desired columns
def getcolumns(bars):
    categories=['address','categories','city','is_open', 'latitude', 'longitude', 'name', 'review_count','stars', 'Monday open', 'Monday close','Tuesday open', 'Tuesday close', 'Wednesday open', 'Wednesday close', 'Thursday open', 'Thursday close', 'Friday open', 'Friday close', 'Saturday open', 'Saturday close', 'Sunday open', 'Sunday close', 'Alcohol', 'WiFi']
    bars2=bars[categories]
    return bars2

def createcsv(bars):
    bars.to_csv('data/torontobars.csv')
```
