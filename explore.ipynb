# Bar Crawl Profiling

### Import required libraries

```python
import numpy as np
import pandas as pd
```

### Import dataset

```python
df = pd.read_json('data/business.json', lines=True)
```

```python
df['categories'] = df['categories'].astype('str') 
```

```python
df
```

```python
# Only keep bars
df_bars = df[(df['categories'].str.contains('Bars') 
             | df['categories'].str.contains('Nightlife') 
             | df['categories'].str.contains('Pubs') )
             & ( ~df['categories'].str.contains('Sushi Bars') 
             & ~df['categories'].str.contains('Juice Bars')) ]
```

```python
df_bars[['city','address']].groupby('city').count().sort_values('address', ascending=False)
```

```python
df[['city','address']].groupby('city').count().sort_values('address', ascending=False)
```

```python
df.to_csv('data/business.csv')
```
