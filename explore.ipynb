```python
import matplotlib.pyplot as plt
import pandas as pd
import calendar
# Load data
columns = ['year', 'month', 'day', 'high_temp', 'low_temp', 'avg_temp']
temps_df = pd.read_csv('electricbill/cambridge_avg_temps.tsv', sep='\t', header=None, names=columns)
temps_df['date'] = pd.to_datetime(temps_df[['year', 'month', 'day']])

columns = ['year', 'month', 'usage']
bills_df = pd.read_csv('electricbill/cambridge_bills.tsv', sep='\t', header=None, names=columns)

temps_gp = temps_df.groupby(['year','month'])['avg_temp'].mean().reset_index()
temps_gp['avg_temp'] = temps_gp['avg_temp'].apply(lambda f: round(f,1))

temps_join = pd.merge(temps_gp, bills_df, on=['year', 'month'])
temps_join.to_csv('electricbill/cambridge_temps_bills.tsv', sep='\t', index=False, header=False)
```

```python
def sample_mean(values):
    return sum(values)/len(values)

def sample_var(values):
    mean = sample_mean(values)
    return 1/(len(values)-1) * sum([(a - mean)**2 for a in values])

def sample_std(values):
    return sample_var(values)**0.5

def sample_cov(x_values, y_values):
    if len(x_values) != len(y_values):
        return None
    x_mean = sample_mean(x_values)
    y_mean = sample_mean(y_values)
    return 1/(len(x_values)-1) * sum([(a - x_mean)*(b - y_mean) for a,b in zip(x_values, y_values)])


def pearson_r(x_values, y_values):
    return sample_cov(x_values, y_values)/(sample_std(x_values)*sample_std(y_values))

def linear_regression(x_values, y_values):
    m = sample_cov(x_values, y_values)/sample_var(x_values)
    b = sample_mean(y_values) - m*sample_mean(x_values)
    
    return m, b, pearson_r(x_values, y_values)
```

```python
temps_join_train = temps_join.loc[lambda f: f['year']<2014]
temps_join_test = temps_join.loc[lambda f: f['year']>=2014]
```

```python
m,b, p = linear_regression(temps_join_train['avg_temp'], temps_join_train['usage'])
```

```python
temps_join_train['predict']= temps_join_train['avg_temp']*m+b
temps_join_test['predict']= temps_join_test['avg_temp']*m+b
```

```python
pearson_r(temps_join_test['usage'],temps_join_test['predict'])
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import calendar
# Load data
columns = ['year', 'month', 'day', 'high_temp', 'low_temp', 'avg_temp']
temps_df = pd.read_csv('electricbill/cambridge_avg_temps.tsv', sep='\t', header=None, names=columns)
temps_df['date'] = pd.to_datetime(temps_df[['year', 'month', 'day']])
temps_df['hot'] = (temps_df['low_temp'] > 65).astype(int)
temps_df['cold'] = (temps_df['high_temp'] <40 ).astype(int)
temps_df['neutral'] = 1 - temps_df['hot'] - temps_df['cold']
```

```python

columns = ['year', 'month', 'usage']
bills_df = pd.read_csv('electricbill/cambridge_bills.tsv', sep='\t', header=None, names=columns)

aggs = { 'avg_temp': 'mean' , 'hot': 'sum', 'cold': 'sum', 'neutral': 'sum'}

temps_gp = temps_df.groupby(['year','month']).agg(aggs).reset_index()
temps_gp['avg_temp'] = temps_gp['avg_temp'].apply(lambda f: round(f,1))

temps_join = pd.merge(temps_gp, bills_df, on=['year', 'month'])
temps_join.to_csv('electricbill/cambridge_temps_bills_2.tsv', sep='\t', index=False, header=False)
```

```python
temps_join_train = temps_join.loc[lambda f: f['year']<2014]
temps_join_test = temps_join.loc[lambda f: f['year']>=2014]
```

```python
from sklearn import datasets, linear_model
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(temps_join_train[['avg_temp', 'hot','cold','neutral']], temps_join_train['usage'])
```

```python
pearson_r(temps_join_test['usage'],regr.predict(temps_join_test[['avg_temp', 'hot','cold','neutral']]))
```
